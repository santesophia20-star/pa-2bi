```python
#3
#numeros pares, o usuario digita um numero e o programa imprimi os numeros pares

numero = int(input("digite um numero positivo:"))
print(f"Selecionando números pares até {numero}:")

for i in range(numero + 1):
    if i % 2 == 0:
        print(i)
```

```python

# 5 - soma, o usuario digita um numero e o programa soma os numeros do zero até o numero

n = int(input("Digite um número: "))

soma = 0
for i in range(n + 1):
    soma += i

print(f"A soma dos números de 0 até {n} é: {soma}")
```

```python
# 6- tabuada, o usuario digita um numero e o programa imprimi a tabuada do numero

a = int(input("Digite um número para ver a tabuada: "))

for i in range(1, 11):
    print(f"{a} x {i} = {a * i}")
```

```python
# 8 - contar vogais, o usuario digita uma palavra e oprograma conta as vogais

palavra = input("Digite uma palavra: ")
vogais = "aeiouAEIOU"
contador_vogais = 0

for letra in palavra:
    if letra in vogais:
        contador_vogais += 1
        print(f"a palavra  '{palavra}' tem {contador_vogais}vogais")
```

```python
# 10 - o usuário digita um número entre 01 e 100 e o programa mostra quais são divisiveis por 3 e 5
numero_limite = int(input("Digite um número entre 1 e 100: "))

if 1 <= numero_limite <= 100:
    print(f"Números divisíveis por 3 e 5 até {numero_limite}:")
    for i in range(1, numero_limite + 1):
        if i % 3 == 0 and i % 5 == 0:
            print(i)
else:
    print("Por favor, digite um número entre 1 e 100.")
```
