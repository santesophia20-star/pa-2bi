```python
numb = int  (input("Digite um numero inteiro positivo: "))
while numb < 0:
  numb = int  (input("Por favor, digite um numero inteiro positivo: "))
fatorial = 1
for i in range (1,numb+1):
  fatorial = fatorial * i
print(f'Fatorial = {fatorial}')
print('Pois:')
print(f'{numb}! = ',end='')
for i in range (numb,1,-1):
  print(f'{i} × ',end='')
print('1')
```

#questão 05

```python
#questão 05
valor_saque = int(input("Digite o valor do saque: "))

notas = [100, 50, 20, 10, 5, 2, 1]

print("\nSaída esperada:\n")

for nota in notas:
    quantidade = valor_saque // nota
    if quantidade > 0:
        print(f"{quantidade} nota(s) de R${nota}")
    valor_saque %= nota
```
